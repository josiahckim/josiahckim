### Initialization 

```python
# Import Packages
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
import math

# Import Data
df = pd.read_csv("Data/winequality-red.csv", sep = ";",na_values="?") # ; is delimiter
print("Number of records: {}\nNumber of variables: {}".format(df.shape[0], df.shape[1])) 

#Take a Peek
df.head(10)
```

### a. Standardize

```python
# Separate 
X, Y = df.iloc[:, :-1], df.iloc[:, -1]

# Min-max Normalization 
x = X.values 
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
col_names = list(df[:0])
X_scaled = pd.DataFrame(x_scaled, columns=col_names[:-1])

final_dataset = pd.concat([X_scaled, Y], axis=1, sort=False)
final_dataset.head(10)
```

### b. Implement Linear Regression

```python
# Separate the response variable 
x_value = X
y_value = final_dataset["quality"]

# Convert to arrays
x = x_value.to_numpy()
y = y_value.to_numpy()

# Model
model = LinearRegression(fit_intercept = True)
model.fit(x, y)

# Predict
y_predicted = model.predict(x)
final_dataset["quality_predicted"] = y_predicted

# RMSE
mse =  mean_squared_error(final_dataset["quality"], final_dataset["quality_predicted"])
rmse = math.sqrt(mse)
print("The RMSE score:", rmse)
```

The RMSE score: 0.6455750670692045

### c. Most important features

```python
# Get coefficients
coefficients = model.coef_
# Create a dictionary with feature names and the respective coefficients
features = {}
for j, i in enumerate(coefficients):
    features[final_dataset.columns.values[j]] = i
# Sort
dict(sorted(features.items(), key=lambda item: item[1]))
```

The most important features for prediction are sulphates, alcohol, and fixed acidity, respectively. 

### d. Linear Regression with Regularization

```python
# Model with Regularization 
model2 = Ridge(alpha=100.0)
model2.fit(x, y)

#Predict
quality_predicted2 = model2.predict(x)
final_dataset['quality_predicted2'] = quality_predicted2
final_dataset.head(10)

# RMSE
mse =  mean_squared_error(final_dataset["quality"], final_dataset["quality_predicted2"])
rmse = math.sqrt(mse)
print("The RMSE score:", rmse)
```

The RMSE score: 0.6693299908684548
