```python
'''
Import statements.
'''
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''
Import dataset and take a peek.
'''
data = datasets.load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target
df
```

## Start Visualizations

```python
'''
Let's look at the distribution of our output variable.
We want a balanced dataset for modelling.
'''
plt.hist(df["target"])
plt.title("Distribution of Output Values")
plt.xlabel("Value")
plt.ylabel("Count")
plt.xticks([0, 1, 2])
```
