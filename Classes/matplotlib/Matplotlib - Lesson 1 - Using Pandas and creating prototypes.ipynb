## Filtering

Both Pandas and D3.js have a type of filtering that are very similar. I will discuss the Pandas version. 

#### Filter by year

```python
data[data.YEAR==2020]
```

#### Filter by year and month

```python
data[(data.YEAR==2020) & (data.MONTH==2)]
```

#### Filter by WT_HAIL or WT_HighWinds

```python
data[(data.WT_HAIL==1) | (data.WT_HIGHWINDS==1)]
```

### Correlation

```python
import matplotlib.pyplot as plt
snowdays = data[(data.SNOW>0)]

plt.scatter(snowdays.TMIN, snowdays.SNOW)
plt.show()
```
